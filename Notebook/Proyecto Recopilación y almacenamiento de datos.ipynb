```python
sns.set_theme()
plt.figure(figsize=(10, 4))
weather_trips.query("weather_conditions=='Good'")['duration_seconds'].hist(
    label='Good',
    ax=plt.gca(),
    density=True,
    alpha=0.5,
    bins=15
)
weather_trips.query("weather_conditions=='Bad'")['duration_seconds'].hist(
    label='Bad', 
    ax=plt.gca(), 
    density=True, 
    alpha=0.5,
    bins=15
)

plt.legend(title='Weather Condition')
plt.title('Distribución del tiempo de viaje por condiciones climáticas')
plt.xlabel('Duración en segundos')
plt.ylabel('Frecuencia')
plt.show("svg")
```

```python
prueba= weather_trips.groupby('weather_conditions')['duration_seconds'].mean()
prueba
```

```python
#Extrayendo datos para condiciones buenas
good= weather_trips[weather_trips['weather_conditions'] == 'Good']['duration_seconds']
good
```

```python
#Extrayendo datos para condiciones malas
bad= weather_trips[weather_trips['weather_conditions'] == 'Bad']['duration_seconds']
bad.describe()
```

```python
# Determinanando las varianzas:
print("Varianza para condiciones buenas es:", good.var())
print("Varianza para condiciones malas es:", bad.var())
```

```python
# Prueba de levene para determinar H0 = la varianza de ambas muestras son iguales:
alpha = 0.05

levene_test_variance_median = st.levene(good, bad, center ='median') #Prueba para determinar si las varianzas son iguales según la Mediana

levene_test_variance_mean = st.levene(good, bad, center ='mean')  #Prueba para determinar si las varianzas son iguales según la Media


if (levene_test_variance_median.pvalue < alpha): 
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
    
if (levene_test_variance_mean.pvalue < alpha): 
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Por tanto,  las varianzas no son significativamente diferentes entre sí

### Prueba la hipótesis:

- "La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos".

H0 = La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare **NO** cambia los sábados lluviosos.\
H1 = La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare  cambia los sábados lluviosos




```python
# Prueba de hipótesis:
alpha= 0.05
results= st.ttest_ind(good, bad, equal_var=True)

print("valor p: ",results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else: 
    print("No podemos rechazar la hipótesis nula")
```

Por tanto, afirmamos que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos

### Explica:

- Cómo planteaste las hipótesis nula y alternativa

    La H0 y H1 se formuló de la sgte. manera:\
    H0 = La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare **NO** cambia los sábados lluviosos.\
H1 = La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare  cambia los sábados lluviosos\

Puesto que, la H0 suele expresar la idea de que no hay diferencias entre los grupos, por eso en esta hipótesis expresamos que 
la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia (O es la misma) independientemente de las condiciones climatológicas.
    
    
- Qué criterio usaste para probar las hipótesis y por qué

    Para probar las hipótesis se utilizó el *valor p* para decidir si se acepta o rechaza la H0. Ya que el valor p es la probabilidad de obtener un resultado al menos tan extremo como el que se está considerando suponiendo que la H0 sea correcta.
